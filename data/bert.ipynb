```python
def _read_csv(input_file, quotechar=None):
    """Reads a tab separated value file."""
    dicts = []
    with open(input_file, 'r', encoding='utf-8') as infs:
        for inf in infs:
            inf = inf.strip()
            dicts.append(inf)
    return dicts
```

```python
def bert_input(s1,s2,max_len):
    q=["[CLS]"]
    q.extend("".join(s1))
#     for i in range (max_len-len(s1)):
#         q.extend(["[MASK]"])
    q.extend(["[SEP]"])
    stype=[0]*len(q)
    q.extend("".join(s2))
#     for i in range (max_len-len(s2)):
#         q.extend(["[MASK]"])
    q.extend(["[SEP]"])
    stype2=[1]*(len(q)-len(stype))
    stype.extend(stype2)
    inputs=" ".join(q)
    return q,stype
```

```python
import torch
from pytorch_pretrained_bert import BertTokenizer, BertModel, BertForMaskedLM,BertForSequenceClassification
tokenizer = BertTokenizer.from_pretrained('vocab.txt')
# Load pre-trained model (weights)
model = BertModel.from_pretrained(r'F:\数据\BERT\chinese_L-12_H-768_A-12\pytorch_weight\bert-base-chinese.tar.gz')
model.eval()
# If you have a GPU, put everything on cuda
model.to('cuda')
```

```python
import re
reg=re.compile("\S+")

c=[]
l=[]
for i in range (len(t1)):
    s1=t1[i]
    s2=t2[i]  
    lable=t3[i]
    s1=",".join(reg.findall(s1))
    s2=",".join(reg.findall(s1))
    try:
        tokenized_text,segments_ids=bert_input(s1,s2,max_len)
        indexed_tokens = tokenizer.convert_tokens_to_ids(tokenized_text)
        tokens_tensor = torch.tensor([indexed_tokens])
        segments_tensors = torch.tensor([segments_ids])
        tokens_tensor = tokens_tensor.to('cuda')
        segments_tensors = segments_tensors.to('cuda')
        with torch.no_grad():
            _, bert_vector = model(tokens_tensor, segments_tensors)
        c.append(bert_vector.data.cpu().numpy()[0])
        l.append(lable)
        print(i)
        if(i>1000):break
    except:
        pass
    
```
